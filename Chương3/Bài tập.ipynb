```python
def read_file_to_list(filename):
    data_list = []
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
    parts = [p.strip() for p in text.replace('\n', ',').split(',')]
    for token in parts:
        if token == '':
            continue
        try:
            if '.' in token:
                value = float(token)
            else:
                value = int(token)
            data_list.append(value)
        except ValueError:
            try:
                data_list.append(float(token))
            except ValueError:
                continue
    return data_list


import numpy as np
height = read_file_to_list('heights_1.txt')
weight = read_file_to_list('weights_1.txt')

# Tạo numpy arrays theo yêu cầu
arr_height = np.array(height)
arr_weight = np.array(weight)

```

```python
arr_height_m = arr_height * 0.0254
```

```python
arr_weight_kg = arr_weight * 0.453592
```

```python
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
```

```python
print(arr_weight_kg[50])
```

```python
arr_height_m_100 = arr_height_m[100:111]
```

```python
underweight_players = arr_bmi < 21
print(underweight_players)
```

```python
avg_height = np.mean(arr_height_m)
avg_weight = np.mean(arr_weight_kg)

print("Chiều cao trung bình: ", avg_height)
print("Cân nặng trung bình: ", avg_weight)
```

```python
max_height = np.max(arr_height_m)
max_weight = np.max(arr_weight_kg)

print("Chiều cao lớn nhất: ", max_height)
print("Cân nặng lớn nhất: ", max_weight)
```

```python
min_height = np.min(arr_height_m)
min_weight = np.min(arr_weight_kg)

print("Chiều cao nhỏ nhất: ", min_height)
print("Cân nặng nhỏ nhất: ", min_weight)
```

```python
import numpy as np
import ast

# Doc du lieu tu tap tin baseball_2D.txt
with open('baseball_2D.txt', 'r', encoding='utf-8-sig') as f:
    data_text = f.read()

# Parse du lieu tu dinh dang list
baseball = ast.literal_eval(data_text)
```

```python
# Tạo 2D numpy array np_baseball
np_baseball = np.array(baseball)
print("Kiểu dữ liệu của np_baseball:", np_baseball.dtype)
print("Kích thước của np_baseball:", np_baseball.shape)
```

```python
# In các giá trị của dòng thứ 50
print("\nDòng thứ 50 (index 49) của np_baseball:")
print(np_baseball[49])
```

```python
#Tạo numpy array np_weight với cột 2
np_weight = np_baseball[:, 1]
print("\nnp_weight (cân nặng) - 10 giá trị đầu tiên:")
print(np_weight[:10])
print(f"Kiểu dữ liệu của np_weight: {np_weight.dtype}")
```

```python
# Chiều cao của vận động viên thứ 124
np_height = np_baseball[:, 0]
height_124 = np_height[123]  # Index 123 vì đếm từ 0
print(f"\nChiều cao của vận động viên thứ 124: {height_124} inches")
```

```python
# Chiều cao trung bình, cân nặng trung bình
avg_height_baseball = np.mean(np_height)
avg_weight_baseball = np.mean(np_weight)

print(f"\nChiều cao trung bình: {avg_height_baseball:.2f} inches")
print(f"Cân nặng trung bình: {avg_weight_baseball:.2f} pounds")
```

```python

# Tính hệ số tương quan Pearson
correlation = np.corrcoef(np_height, np_weight)[0, 1]

print(f"\nHệ số tương quan giữa chiều cao và cân nặng: {correlation:.4f}")

# Phân tích
if correlation > 0.7:
    print("Nhận xét: Có tương quan thuận MẠNH giữa chiều cao và cân nặng.")
    print("Càng cao thì cân nặng có xu hướng càng nặng.")
elif correlation > 0.3:
    print("Nhận xét: Có tương quan thuận trung bình giữa chiều cao và cân nặng.")
    print("Những cầu thủ cao hơn có xu hướng nặng hơn.")
elif correlation > 0:
    print("Nhận xét: Có tương quan thuận yếu giữa chiều cao và cân nặng.")
elif correlation > -0.3:
    print("Nhận xét: Có tương quan nghịch yếu giữa chiều cao và cân nặng.")
else:
    print("Nhận xét: Có tương quan nghịch giữa chiều cao và cân nặng.")
```

```python
import numpy as np

# Đọc positions.txt (thường không có BOM ở đầu, nhưng vẫn làm sạch)
with open('positions.txt', 'r', encoding='utf-8-sig') as f:
    positions = [pos.strip() for pos in f.read().strip().split(',') if pos.strip()]

# Đọc heights.txt — quan trọng: dùng encoding='utf-8-sig' để tự động bỏ BOM
with open('heights.txt', 'r', encoding='utf-8-sig') as f:
    raw_text = f.read().strip()
    height_strings = [h.strip() for h in raw_text.split(',') if h.strip()]
    heights = [int(h) for h in height_strings]  # giờ sẽ không còn 'ï»¿191'

# Kiểm tra độ dài (phải bằng nhau)
print(f"Số cầu thủ: {len(positions)}")
print(f"Số chiều cao: {len(heights)}")
assert len(positions) == len(heights), "Lỗi: 2 file có số lượng dữ liệu khác nhau!"

# ==================================================================
# 1a) Tạo np_positions
# ==================================================================
np_positions = np.array(positions)
print("\n1a) np_positions dtype:", np_positions.dtype)

# ==================================================================
# 1b) Tạo np_heights
# ==================================================================
np_heights = np.array(heights, dtype=float)  # dùng float để sau tính mean chính xác hơn
print("1b) np_heights dtype:", np_heights.dtype)

# ==================================================================
# 2) Chiều cao trung bình của GK
# ==================================================================
gk_heights = np_heights[np_positions == 'GK']
avg_gk = gk_heights.mean()
print(f"\n2) Chiều cao trung bình thủ môn (GK): {avg_gk:.2f} cm (n = {len(gk_heights)})")

# ==================================================================
# 3) Chiều cao trung bình các vị trí khác
# ==================================================================
other_heights = np_heights[np_positions != 'GK']
avg_other = other_heights.mean()
print(f"3) Chiều cao trung bình các vị trí khác: {avg_other:.2f} cm (n = {len(other_heights)})")

# ==================================================================
# 4) Tạo structured array players
# ==================================================================
dt = np.dtype([
    ('position', 'U5'),
    ('height', float)
])
players = np.array(list(zip(positions, heights)), dtype=dt)
print("\n4) Structured array 'players' đã tạo thành công!")

# ==================================================================
# 5) Sắp xếp và tìm cầu thủ cao nhất/thấp nhất
# ==================================================================
players_sorted = np.sort(players, order='height')

lowest = players_sorted[0]
highest = players_sorted[-1]

print(f"\n5) Kết quả sau khi sắp xếp theo chiều cao:")
print(f"   → Thấp nhất: {lowest['position']} - {lowest['height']:.0f} cm")
print(f"   → Cao nhất : {highest['position']} - {highest['height']:.0f} cm")

# Bonus: Top 5 cao nhất & thấp nhất
print("\n   Top 5 cầu thủ cao nhất:")
for p in players_sorted[-5:][::-1]:
    print(f"   • {p['position']:2} : {p['height']:.0f} cm")

print("   Top 5 cầu thủ thấp nhất:")
for p in players_sorted[:5]:
    print(f"   • {p['position']:2} : {p['height']:.0f} cm")
```

```python
import pandas as pd


drink = pd.read_csv('drinks.csv', index_col=0)

print("Kiểu dữ liệu của drink:", type(drink))
print("Kích thước của drink:", drink.shape)
print("Tên các cột của drink:", drink.columns)
print("5 dòng dữ liệu đầu tiên của drink:\n", drink.head())
print("5 dòng dữ liệu cuối cùng của drink:\n", drink.tail())
```

```python
average_beer = drink.groupby('continent')['beer_servings'].mean()
print("Số lượng bia tiêu thụ trung bình ở mỗi châu lục:\n", average_beer)
```

```python
wine_stats = drink.groupby('continent')['wine_servings'].describe()
print("Thông tin thống kê tổng quát về số lượng rượu vang được tiêu thụ ở mỗi châu lục:\n", wine_stats)
```

```python
mean_beer_wine = drink.groupby('continent')[['beer_servings', 'wine_servings']].mean()
print("Số lượng các loại bia và rượu tiêu thụ trung bình ở mỗi châu lục:\n", mean_beer_wine)
```

```python
median_beer_wine = drink.groupby('continent')[['beer_servings', 'wine_servings']].median()
print("Giá trị trung vị cho các loại bia và rượu tiêu thụ ở mỗi châu lục:\n", median_beer_wine)
```

```python
spirit_stats = drink.groupby('continent')['spirit_servings'].aggregate(['mean', 'max', 'min'])
print("Số lượng rượu mạnh tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục:\n", spirit_stats)
```

```python
sorted_beer = drink.sort_values(by='beer_servings', ascending=False)
top_5_beer_consumers = sorted_beer.head(5)
bottom_5_beer_consumers = sorted_beer.tail(5)
print("5 quốc gia có lượng tiêu thụ bia nhiều nhất:\n", top_5_beer_consumers[['beer_servings']])
print("5 quốc gia có lượng tiêu thụ bia ít nhất:\n", bottom_5_beer_consumers[['beer_servings']])
```
