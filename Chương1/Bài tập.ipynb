```python
class HinhChuNhat:
    
    def __init__(self):
        self.chieu_dai = 0
        self.chieu_rong = 0

    def nhap(self):
        self.chieu_dai = float(input("  Nhập chiều dài: "))
        self.chieu_rong = float(input("  Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return (self.chieu_dai + self.chieu_rong) * 2

    def tinh_dien_tich(self):
        return self.chieu_dai * self.chieu_rong

    def in_thong_tin(self):
        chu_vi = self.tinh_chu_vi()
        dien_tich = self.tinh_dien_tich()
        
        print("\n THÔNG TIN HÌNH CHỮ NHẬT ")
        print(f"Chiều dài  : {self.chieu_dai}")
        print(f"Chiều rộng : {self.chieu_rong}")
        print(f"Chu vi     : {chu_vi}")
        print(f"Diện tích  : {dien_tich}")

# Tạo một đối tượng (instance) từ lớp HinhChuNhat
hcn_cua_toi = HinhChuNhat()

# Yêu cầu đối tượng tự nhập dữ liệu
hcn_cua_toi.nhap()

# Yêu cầu đối tượng in thông tin của nó ra màn hình
hcn_cua_toi.in_thong_tin()
```

```python
class TS:
    def __init__(self):
        self.ho_ten = ""
        self.diem_toan = 0.0
        self.diem_ly = 0.0
        self.diem_hoa = 0.0

    def nhap(self):
        self.ho_ten = input("  Nhập họ tên thí sinh: ")
        self.diem_toan = float(input("  Nhập điểm Toán: "))
        self.diem_ly = float(input("  Nhập điểm Lý: "))
        self.diem_hoa = float(input("  Nhập điểm Hoá: "))

    def tinh_tong_diem(self):
        return self.diem_toan + self.diem_ly + self.diem_hoa

    def in_thong_tin(self):
        tong_diem = self.tinh_tong_diem()
        print(f"Họ tên: {self.ho_ten:<25} | "
              f"Toán: {self.diem_toan:<4} | "
              f"Lý: {self.diem_ly:<4} | "
              f"Hoá: {self.diem_hoa:<4} | "
              f"TỔNG ĐIỂM: {tong_diem:<5}")


# Định nghĩa điểm chuẩn
DIEM_CHUAN = 20.0

# 1. Nhập danh sách thí sinh
print("BẮT ĐẦU NHẬP DANH SÁCH THÍ SINH")
so_luong = int(input("Nhập số lượng thí sinh: "))

# Khởi tạo một danh sách (list) rỗng để lưu các đối tượng TS
danh_sach_ts = []

for i in range(so_luong):

    # Tạo một đối tượng Thí Sinh mới
    thi_sinh = TS()
    
    # Yêu cầu đối tượng đó tự nhập thông tin
    thi_sinh.nhap()
    
    # Thêm đối tượng đã có thông tin vào danh sách
    danh_sach_ts.append(thi_sinh)

print("\nĐã nhập xong danh sách!")


# 2. Xử lý và in danh sách trúng tuyển

danh_sach_trung_tuyen = [
    ts for ts in danh_sach_ts if ts.tinh_tong_diem() >= DIEM_CHUAN
]

danh_sach_trung_tuyen.sort(key=lambda ts: ts.tinh_tong_diem(), reverse=True)


# 2c. In kết quả ra màn hình
print("\n" + "="*60)
print(f"DANH SÁCH THÍ SINH TRÚNG TUYỂN (Điểm chuẩn = {DIEM_CHUAN})")
print("Sắp xếp theo thứ tự tổng điểm từ cao xuống thấp")
print("="*60)

if not danh_sach_trung_tuyen:
    print("Không có thí sinh nào trúng tuyển.")
else:
    # Duyệt qua danh sách đã lọc và sắp xếp
    for (i, ts) in enumerate(danh_sach_trung_tuyen):
        print(f"{i + 1:02}. ", end="") # In số thứ tự
        ts.in_thong_tin() # Yêu cầu đối tượng tự in thông tin

print("="*60)
```

```python
class PS:
    def __init__(self):
        self.tu = 0    
        self.mau = 1   

    # Kiểm tra tính hợp lệ của phân số (mẫu số khác 0)
    def hop_le(self):
        return self.mau != 0

    # Nhập phân số từ bàn phím
    def nhap(self):
        while True:
            try:
                self.tu = int(input("Nhập tử số: "))
                self.mau = int(input("Nhập mẫu số: "))
                
                if not self.hop_le():
                    print("Lỗi: Mẫu số không được bằng 0! Vui lòng nhập lại.\n")
                else:
                    break
            except ValueError:
                print("Vui lòng nhập số nguyên hợp lệ!\n")

    # In phân số ra màn hình
    def in_phan_so(self):
        if self.tu == 0:
            print("0")
        elif self.mau == 1:
            print(self.tu)
        else:
            print(f"{self.tu}/{self.mau}")


# Chương trình chính
def main():
    print("NHẬP PHÂN SỐ")
    ps = PS()
    ps.nhap()
    
    print("\nPhân số vừa nhập:", end=" ")
    ps.in_phan_so()


if __name__ == "__main__":
    main()
```

```python
class Stack:
    
    def __init__(self, max_size=10):
        # Dữ liệu là mảng (list) chứa các số float
        self.items = []
        self.max_size = max_size
        print(f"Đã tạo Stack với kích thước tối đa là {self.max_size}")


    def isEmpty(self):
        # Nếu độ dài (len) của list là 0, tức là rỗng
        return len(self.items) == 0

    def isFull(self):
        """Kiểm tra ngăn xếp đã đầy chưa (dựa trên max_size)"""
        return len(self.items) == self.max_size

    def push(self, item):
        if self.isFull():
            print(f"Lỗi: Ngăn xếp đã đầy! Không thể thêm {item}.")
        else:
            # Chuyển đổi item sang float theo yêu cầu
            data = float(item) 
            self.items.append(data)
            print(f"Đã PUSH [ {data} ] vào ngăn xếp.")

    def pop(self):
        if self.isEmpty():
            print("Lỗi: Ngăn xếp rỗng! Không thể POP.")
            return None
        else:
            # list.pop() của Python trả về và xoá phần tử cuối cùng
            item = self.items.pop()
            print(f"Đã POP [ {item} ] ra khỏi ngăn xếp.")
            return item

    def count(self):
        # Chỉ cần trả về độ dài của list
        return len(self.items)

    def print(self):
        """In nội dung của ngăn xếp từ đỉnh xuống đáy"""
        if self.isEmpty():
            print("Nội dung Stack: [ RỖNG ]")
            return
        
        print("\n--- NỘI DUNG STACK ---")
        print("  ĐỈNH (TOP)")
        for item in reversed(self.items):
            print(f"   [ {item} ]")
        print("  ĐÁY (BOTTOM)")
        print("-----------------------")
```

```python
# Tạo một ngăn xếp có kích thước tối đa là 3 phần tử
s = Stack(max_size=3)

# Kiểm tra lúc mới tạo
print(f"Stack có rỗng không? {s.isEmpty()}")
print(f"Số phần tử: {s.count()}")
s.print()

print("\n--- Bắt đầu PUSH ---")
# Thêm các phần tử
s.push(1.1)
s.push(2.2)
s.push(3.3)

# Kiểm tra khi đã đầy
print("\n--- Kiểm tra khi đầy ---")
print(f"Stack có đầy không? {s.isFull()}")
print(f"Số phần tử: {s.count()}")
s.print()

# Thử PUSH khi đã đầy
print("\n--- Thử PUSH khi đầy ---")
s.push(4.4) # Sẽ báo lỗi

# Bắt đầu POP
print("\n--- Bắt đầu POP ---")
s.pop() # Lấy 3.3 ra
s.pop() # Lấy 2.2 ra

# Kiểm tra sau khi POP
print("\n--- Kiểm tra sau khi POP ---")
print(f"Stack có đầy không? {s.isFull()}")
print(f"Số phần tử: {s.count()}")
s.print() # Chỉ còn [1.1]

# POP phần tử cuối cùng
print("\n--- POP lần cuối ---")
s.pop() # Lấy 1.1 ra

# Kiểm tra khi đã rỗng
print("\n--- Kiểm tra khi rỗng ---")
print(f"Stack có rỗng không? {s.isEmpty()}")
s.print()

# Thử POP khi đã rỗng
print("\n--- Thử POP khi rỗng ---")
s.pop() # Sẽ báo lỗi
```

```python
class Date:


    def __init__(self, day=1, month=1, year=2025):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")

    # --- Các phương thức hỗ trợ cho hàm next() ---

    def _is_leap_year(self, year):
        return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

    def _get_days_in_month(self, month, year):
        if month == 2:
            # Nếu là tháng 2, kiểm tra năm nhuận
            return 29 if self._is_leap_year(year) else 28
        elif month in [4, 6, 9, 11]:
            # Các tháng 4, 6, 9, 11 có 30 ngày
            return 30
        else:
            # Các tháng còn lại (1, 3, 5, 7, 8, 10, 12) có 31 ngày
            return 31

    # --- Phương thức chính ---

    def next(self):
        days_in_current_month = self._get_days_in_month(self.month, self.year)

        # Kiểm tra xem ngày hiện tại có phải là ngày cuối tháng không
        if self.day < days_in_current_month:
            self.day += 1
        else:
            self.day = 1
            
            if self.month == 12:
                self.month = 1
                self.year += 1
            else:
                self.month += 1


# --- Ví dụ sử dụng lớp Date ---


print("\n--- Thử ngày bình thường ---")
d2 = Date(15, 10, 2025)
d2.display()
d2.next()
print("Ngày tiếp theo là:")
d2.display() # 16/10/2025

print("\n--- Thử ngày cuối tháng (tháng có 30 ngày) ---")
d3 = Date(30, 4, 2024)
d3.display()
d3.next()
print("Ngày tiếp theo là:")
d3.display() # 01/05/2024

print("\n--- Thử ngày cuối năm ---")
d4 = Date(31, 12, 2025)
d4.display()
d4.next()
print("Ngày tiếp theo là:")
d4.display() # 01/01/2026

print("\n--- Thử năm nhuận (2024 là năm nhuận) ---")
d5 = Date(28, 2, 2024)
d5.display()
d5.next()
print("Ngày tiếp theo là:")
d5.display() # 29/02/2024
d5.next()
print("Ngày tiếp theo nữa là:")
d5.display() # 01/03/2024
```

```python
import math
from abc import ABC, abstractmethod


class DaGiac(ABC):

    def __init__(self):
        super().__init__()

    @abstractmethod
    def chu_vi(self):
        pass

    @abstractmethod
    def dien_tich(self):
        pass

    def in_thong_tin(self):
        print(f"  - Chu vi: {self.chu_vi():.2f}")
        print(f"  - Diện tích: {self.dien_tich():.2f}")



class HinhBinhHanh(DaGiac):
    def __init__(self, a, b, h):
        super().__init__()
        self.a = a
        self.b = b
        self.h = h # Chiều cao tương ứng với cạnh a
        print(f"Đã tạo Hình Bình Hành (a={a}, b={b}, h={h})")

    def chu_vi(self):
        return 2 * (self.a + self.b)

    def dien_tich(self):
        return self.a * self.h



class HinhChuNhat(HinhBinhHanh):
    def __init__(self, chieu_dai, chieu_rong):
        super().__init__(a=chieu_dai, b=chieu_rong, h=chieu_rong)
        print("-> Đây cũng là một Hình Chữ Nhật")

    def dien_tich(self):
        return self.a * self.b



class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(chieu_dai=canh, chieu_rong=canh)
        self.canh = canh
        print("-> Đây cũng là một Hình Vuông")

# CHƯƠNG TRÌNH CHÍNH

def nhap_so_thuc(thong_bao):
    """Hàm tiện ích để nhập số thực, chống lỗi."""
    while True:
        try:
            so = float(input(thong_bao))
            if so <= 0:
                print("Lỗi: Vui lòng nhập một số dương.")
            else:
                return so
        except ValueError:
            print("Lỗi: Vui lòng nhập đúng định dạng số.")

def main():
    while True:

        
        lua_chon = input("Bạn muốn nhập hình nào? (0-3): ")

        hinh = None

        if lua_chon == '1':
            print("\n** Nhập thông tin Hình Bình Hành **")
            a = nhap_so_thuc("Nhập độ dài cạnh đáy (a): ")
            b = nhap_so_thuc("Nhập độ dài cạnh bên (b): ")
            h = nhap_so_thuc(f"Nhập chiều cao (h) tương ứng với cạnh {a}: ")
            hinh = HinhBinhHanh(a, b, h)

        elif lua_chon == '2':
            print("\n** Nhập thông tin Hình Chữ Nhật **")
            dai = nhap_so_thuc("Nhập chiều dài: ")
            rong = nhap_so_thuc("Nhập chiều rộng: ")
            hinh = HinhChuNhat(dai, rong)

        elif lua_chon == '3':
            print("\n** Nhập thông tin Hình Vuông **")
            canh = nhap_so_thuc("Nhập độ dài cạnh: ")
            hinh = HinhVuong(canh)

        elif lua_chon == '0':
            break

        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn lại.")
            continue

        # Nếu đã tạo hình thành công, in kết quả
        if hinh:
            print("\n** Kết quả **")
            hinh.in_thong_tin()

if __name__ == "__main__":
    main()
```

```python
import math

class Diem:
    def __init__(self, x=0, y=0):
        self._x = x 
        self._y = y

    def nhap(self):
        try:
            self._x = float(input("Nhap toa do tam x: "))
            self._y = float(input("Nhap toa do tam y: "))
        except ValueError:
            print("Toa do khong hop le, dat mac dinh la (0, 0)")
            self._x = 0
            self._y = 0

    def xuat(self):
        """Hàm xuất tọa độ"""
        print(f"Tam tai: ({self._x}, {self._y})")

class Elip(Diem):
    def __init__(self, x=0, y=0, ban_kinh_lon=0, ban_kinh_nho=0):
        # Gọi constructor của lớp cha (Diem)
        super().__init__(x, y)
        self._ban_kinh_lon = ban_kinh_lon
        self._ban_kinh_nho = ban_kinh_nho

    def nhap(self):
        # Gọi hàm nhap() của lớp cha (Diem) để nhập tâm
        super().nhap()
        try:
            self._ban_kinh_lon = float(input("Nhap ban kinh truc lon (a): "))
            self._ban_kinh_nho = float(input("Nhap ban kinh truc nho (b): "))
            
            if self._ban_kinh_nho > self._ban_kinh_lon:
                print("Ban kinh nho (b) dang lon hon ban kinh lon (a). Tu dong hoan doi...")
                # Hoán đổi giá trị
                self._ban_kinh_lon, self._ban_kinh_nho = self._ban_kinh_nho, self._ban_kinh_lon

        except ValueError:
            print("Ban kinh khong hop le, dat mac dinh la 0")
            self._ban_kinh_lon = 0
            self._ban_kinh_nho = 0

    def xuat(self):
        super().xuat() # Gọi hàm xuat() của lớp Diem để xuất tâm
        print(f"Ban kinh lon (a): {self._ban_kinh_lon}")
        print(f"Ban kinh nho (b): {self._ban_kinh_nho}")

    def tinh_dien_tich(self):
        return math.pi * self._ban_kinh_lon * self._ban_kinh_nho

class DuongTron(Elip):
    def __init__(self, x=0, y=0, r=0):
        super().__init__(x, y, r, r)

    def nhap(self):
        """Hàm nhập cho DuongTron"""
        # Chỉ cần nhập tâm và 1 bán kính r
        super(Elip, self).nhap() # Gọi hàm nhap() của lớp Diem
        try:
            r = float(input("Nhap ban kinh (r): "))
            # Gán r cho cả hai thuộc tính kế thừa từ Elip
            self._ban_kinh_lon = r
            self._ban_kinh_nho = r
        except ValueError:
            print("Ban kinh khong hop le, dat mac dinh la 0")
            self._ban_kinh_lon = 0
            self._ban_kinh_nho = 0

    def xuat(self):
        """Hàm xuất cho DuongTron"""
        super(Elip, self).xuat() 
        print(f"Ban kinh (r): {self._ban_kinh_lon}")




if __name__ == "__main__":
    e = Elip() # Khai báo một đối tượng Elip
    e.nhap() # Yêu cầu người dùng nhập thông tin

    print("\n--- THONG TIN ELIP VUA NHAP ---")
    e.xuat() # Hiển thị lại thông tin đã nhập

    # Tính và in diện tích
    dien_tich = e.tinh_dien_tich()
    print(f"\n==> Dien tich cua Elip la: {dien_tich:.6f}") # Làm tròn 6 chữ số
```

```python
import math

class TamGiac:
    def __init__(self, a, b, c):
        if not self._la_tam_giac_hop_le(a, b, c):
            raise ValueError("Ba cạnh không tạo thành tam giác hợp lệ")
        
        self.a = a
        self.b = b
        self.c = c
    
    def _la_tam_giac_hop_le(self, a, b, c):
        return (a + b > c) and (a + c > b) and (b + c > a) and (a > 0) and (b > 0) and (c > 0)
    
    def chu_vi(self):

        return self.a + self.b + self.c
    
    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
    
    def __str__(self):
        return f"Tam giác có 3 cạnh: {self.a}, {self.b}, {self.c}"


class TamGiacVuong(TamGiac):
    def __init__(self, canh_goc_vuong1, canh_goc_vuong2):

        canh_huyen = math.sqrt(canh_goc_vuong1**2 + canh_goc_vuong2**2)
        super().__init__(canh_goc_vuong1, canh_goc_vuong2, canh_huyen)
        self.canh_goc_vuong1 = canh_goc_vuong1
        self.canh_goc_vuong2 = canh_goc_vuong2
        self.canh_huyen = canh_huyen
    
    def _la_tam_giac_vuong(self):
        """Kiểm tra tam giác vuông theo định lý Pythagoras"""
        return abs(self.a**2 + self.b**2 - self.c**2) < 1e-10
    
    def __str__(self):
        return f"Tam giác vuông có 2 cạnh góc vuông: {self.canh_goc_vuong1}, {self.canh_goc_vuong2} và cạnh huyền: {self.canh_huyen:.2f}"


class TamGiacCan(TamGiac):
    def __init__(self, canh_ben, canh_day):
        """
        Khởi tạo tam giác cân với cạnh bên và cạnh đáy
        """
        super().__init__(canh_ben, canh_ben, canh_day)
        self.canh_ben = canh_ben
        self.canh_day = canh_day
    
    def _la_tam_giac_can(self):
        """Kiểm tra tam giác cân"""
        return (self.a == self.b) or (self.a == self.c) or (self.b == self.c)
    
    def __str__(self):
        return f"Tam giác cân có cạnh bên: {self.canh_ben}, cạnh đáy: {self.canh_day}"


class TamGiacDeu(TamGiacCan):
    def __init__(self, canh):
        """
        Khởi tạo tam giác đều với độ dài cạnh
        """
        super().__init__(canh, canh)
        self.canh = canh
    
    def _la_tam_giac_deu(self):
        """Kiểm tra tam giác đều"""
        return self.a == self.b == self.c
    
    def __str__(self):
        return f"Tam giác đều có cạnh: {self.canh}"


# Chương trình chính để kiểm tra các lớp
if __name__ == "__main__":
    print("=== KIỂM TRA CÁC LỚP TAM GIÁC ===\n")
    
    try:
        # Tạo và kiểm tra tam giác thường
        tg = TamGiac(3, 4, 5)
        print(tg)
        print(f"Chu vi: {tg.chu_vi()}")
        print(f"Diện tích: {tg.dien_tich():.2f}")
        print()
        
        # Tạo và kiểm tra tam giác vuông
        tgv = TamGiacVuong(3, 4)
        print(tgv)
        print(f"Chu vi: {tgv.chu_vi():.2f}")
        print(f"Diện tích: {tgv.dien_tich():.2f}")
        print()
        
        # Tạo và kiểm tra tam giác cân
        tgc = TamGiacCan(5, 6)
        print(tgc)
        print(f"Chu vi: {tgc.chu_vi()}")
        print(f"Diện tích: {tgc.dien_tich():.2f}")
        print()
        
        # Tạo và kiểm tra tam giác đều
        tgd = TamGiacDeu(5)
        print(tgd)
        print(f"Chu vi: {tgd.chu_vi()}")
        print(f"Diện tích: {tgd.dien_tich():.2f}")
        print()
        
        # Kiểm tra tam giác không hợp lệ
        try:
            tg_khong_hop_le = TamGiac(1, 2, 5)
        except ValueError as e:
            print(f"Lỗi khi tạo tam giác: {e}")
            
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")
```
